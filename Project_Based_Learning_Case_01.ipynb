```python
#Import Libraries
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
#Create DataFrame
Transactions_Amount = pd.DataFrame([[100],
                   [150],
                   [50],
                   [100],
                   [130],
                   [120],
                   [100],
                   [85],
                   [70],
                   [150],
                   [150],
                   [120],
                   [50],
                   [100],
                   [100],
                   [140],
                   [90],
                   [150],
                   [50],
                   [90],
                   [120],
                   [100],
                   [110],
                   [75],
                   [65]],
                  columns = ["Transantions_Amount"])
Transactions_Amount
```

```python
#Central Tendency
Transactions_Amount = (100, 150, 50, 100, 130, 120, 100, 85, 70, 150, 150, 120, 50, 100, 100, 140, 90, 150, 50, 90, 120, 100, 110, 75, 65)

import statistics
print(statistics.mean(Transactions_Amount))
print(statistics.median(Transactions_Amount))
```

```python
#Tulis analisismu dari Central Tendency di sini
#Statistik nilai mean/rata-rata pada transaction amount berada pada nilai 102.6
#statistik nilai median/nilai tengah pada transaction amount berada pada nilai 100
```

```python
#Variability
import numpy as np
print(np.var(Transactions_Amount))
print(np.std(Transactions_Amount))
print(np.max(Transactions_Amount) - np.min(Transactions_Amount))
print(np.quantile(Transactions_Amount, 0.25))
print(np.quantile(Transactions_Amount, 0.5))
print(np.quantile(Transactions_Amount, 0.75))
```

```python
#Tulis analisismu dari Variability di sini
# Statistik nilai standar variansi (var) pada transaction amount berada pada nilai 972.2
# Statistik nilai standar deviasi (std) pada transaction amount berada pada nilai 31.1
# Statistik nilai range (rentang nilai tertinggi dikurang nilai terendah) pada transaction amount berada pada nilai 100
# Statistik nilai persebaran rendah pada transaction amount berada pada nilai 85.0
# Statistik nilai persebaran tengah pada transaction amount berada pada nilai 100.0
# Statistik nilai persebaran tinggi pada transaction amount berada pada nilai 120.0
```

```python
#Define H0 and H1
#H0: Rata-rata transactions amount = 100
#H1: Rata-rata transactions amount > 100
```

```python
#Define alpha = 0.05
```

```python
#Calculate t statistics and P-value
#Perform one sample t-test

stats.ttest_1samp(a=Transactions_Amount, popmean=100)
```

```python
#statistic=0.4085001556802841 , pvalue=0.6865284813438117
```

```python
#Conclude with alpha and p-value
#P-value lebih besar dari alpha (5%), maka gagal menolak H0
#Conclusion: training kepada pekerja sales tidak efektif, rata-rata transactions amount pejualan terbaru masih tetap berada pada 100
```

```python
#Conclude with t statistics and critical region
#Find the critical region
```

```python
#H1: larger (right Slide)
stats.t.ppf(1-0.05, 24)
```

```python
#Critical region result
#Stistik region t > 1.71
#statistik berada diluar statistik region, maka gagal menolak H0
#Conclusion: training kepada pekerja sales tidak efektif, rata-rata transactions amount pejualan terbaru masih tetap berada pada 100
```
